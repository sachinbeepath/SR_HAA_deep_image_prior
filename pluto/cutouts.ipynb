```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
from astropy import wcs
from astropy.nddata.utils import Cutout2D
import os
from astropy.visualization import (ZScaleInterval, ImageNormalize)
```

```python
for file in os.listdir(os.getcwd()):
    if file.endswith('.fit'):
        print(file)
        data, hdr = fits.getdata(file, 0, header=True) #'sci' image[1] data and header
        w = wcs.WCS(hdr)
        pixcrd2 = w.wcs_world2pix([[269.77, -14.105]], 0)
        print(pixcrd2)
        centered_cut = Cutout2D(data, (512, 512), (650, 650))
        #fits.writeto('./cutout_' + file, centered_cut.data, header=hdr, overwrite =True)
        norm = ImageNormalize(centered_cut.data, interval = ZScaleInterval())
        plt.imshow(centered_cut.data,norm=norm)
        plt.show()
        print(centered_cut.data[0].dtype)
    else: 
        continue


    
```
